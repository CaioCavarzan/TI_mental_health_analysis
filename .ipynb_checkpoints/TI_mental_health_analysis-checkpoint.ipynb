#Machine Learning com dados de Saúde Mental em TI

##O presente estudo objetiva realizar o treinamento de um modelo de Machine Learning capaz de prever com acurácia satisfatória a tendência de um profissional de TI buscar ajuda profissional em questões de saúde mental

Columns Content:
- Age
  
- Gender

- Country

- self_employed: Are you self-employed?

- family_history: Do you have a family history of mental illness?

- treatment: Have you sought treatment for a mental health condition?

- work_interfere: If you have a mental health condition, do you feel that it interferes with your work?

- no_employees: How many employees does your company or organization have?

- remote_work: Do you work remotely (outside of an office) at least 50% of the time?

- tech_company: Is your employer primarily a tech company/organization?

- benefits: Does your employer provide mental health benefits?

- care_options: Do you know the options for mental health care your employer provides?

- wellness_program: Has your employer ever discussed mental health as part of an employee wellness program?

- seek_help: Does your employer provide resources to learn more about mental health issues and how to seek help?

- anonymity: Is your anonymity protected if you choose to take advantage of mental health or substance abuse treatment resources?

- leave: How easy is it for you to take medical leave for a mental health condition?

- mental_health_consequence: Do you think that discussing a mental health issue with your employer would have negative consequences?

- phys_health_consequence: Do you think that discussing a physical health issue with your employer would have negative consequences?

- coworkers: Would you be willing to discuss a mental health issue with your coworkers?

- supervisor: Would you be willing to discuss a mental health issue with your direct supervisor(s)?

- mental_health_interview: Would you bring up a mental health issue with a potential employer in an interview?

- phys_health_interview: Would you bring up a physical health issue with a potential employer in an interview?

- mental_vs_physical: Do you feel that your employer takes mental health as seriously as physical health?

- obs_consequence: Have you heard of or observed negative consequences for coworkers with mental health conditions in your workplace?


Importação das libs que serão utilizadas no projeto

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import matplotlib.pyplot as plt
```

Leitura do dataset em .csv com pandas

```python
df = pd.read_csv("survey.csv")
```

1ª Exibição para análise exploratória do dataset

```python
display(df)
```

###Início do tratamento dos dados

1) Remoção das colunas "Timestamp", "state" e "comments" 

```python
df.drop("Timestamp", axis=1, inplace=True)
df.drop("state", axis=1, inplace=True)
df.drop("comments", axis=1, inplace=True)
```

2) Tratamento da idade para valores >= 18 e <= 100

```python
df = df[df["Age"] >= 18]
df = df[df["Age"] <= 100]
```

3) Tratamento da coluna de gêneros

```python
df["Gender"] = df["Gender"].str.strip().str.upper()
valores_troca_MALE = ["M", "MAKE", "MAN", "MALE", "MAL", "MALR", "MAIL", "MSLE", "MAILE", "GUY (-ISH) ^_^","CIS MALE", 
                 "CIS MAN", "MALE LEANING ANDROGYNOUS", "SOMETHING KINDA MALE?", "OSTENSIBLY MALE, UNSURE WHAT THAT REALLY MEANS","MALE-ISH", "MALE (CIS)"]
df["Gender"] = df["Gender"].replace(valores_troca_MALE, "MALE")



valores_troca_FEMALE = ["F", "WOMAN", "FEMALE (TRANS)", "FEMAIL", "FEMALE", "TRANS WOMAN", "TRANS-FEMALE", 
                        "CIS-FEMALE/FEMME", "CIS FEMALE", "FEMAKE","FEMALE (CIS)"]
df["Gender"] = df["Gender"].replace(valores_troca_FEMALE, "FEMALE")



permitidos = ["MALE", "FEMALE"]
df.loc[~df["Gender"].isin(permitidos), "Gender"] = "OTHER"
```

4. Tratamento da coluna "Country"

   Considerando que mais da metade dos entrevistados responderam "United States", manteremos essa categoria e unificaremos todas as outras respostas como "Others"

```python
df["Country"].str.strip()
df.loc[df["Country"] != 'United States', "Country"] = 'Other'
```

5) Tratamento da coluna de número de colaboradores na empresa do entrevistado

```python
df['no_employees'] = df['no_employees'].replace("1-5", "Micro Business")
df['no_employees'] = df['no_employees'].replace("6-25", "Small Business (S1)")
df['no_employees'] = df['no_employees'].replace("26-100", "Small Business (S2)")
df['no_employees'] = df['no_employees'].replace("100-500", "Medium-sized Business")
df['no_employees'] = df['no_employees'].replace("500-1000", "Upper-mid Market")
df['no_employees'] = df['no_employees'].replace("More than 1000", "Large Enterprise")
```

###Tratamento das Strings para valores numéricos

A intenção é habilitar as colunas para aplicação da Matriz de Correlação de Pearson e uso no treinamento do modelo

1) One-Hot Encoding (tratamento de valores nominais) das colunas "Gender" e "Country"

```python
df_processado = pd.get_dummies(df, columns=["Gender", "Country"])

#Removendo espaços do nome da coluna 'Country_United States'
df_processado.rename(columns={'Country_United States' : 'Country_United_States'}, inplace = True)

df_processado["Gender_FEMALE"] = df_processado["Gender_FEMALE"].astype(int)
df_processado["Gender_MALE"] = df_processado["Gender_MALE"].astype(int)
df_processado["Gender_OTHER"] = df_processado["Gender_OTHER"].astype(int)

df_processado["Country_Other"] = df_processado["Country_Other"].astype(int)
df_processado["Country_United_States"] = df_processado["Country_United_States"].astype(int)

display(df_processado)
```

2) Transformação das colunas com dados dicotômicos (sim ou não)

   Colunas: 'self_employed', 'family_history', 'treatment', 'remote_work', 'tech_company', 'obs_consequence'

```python
colunas_dicotomicas = {'self_employed', 'family_history', 'treatment', 'remote_work', 'tech_company', 'obs_consequence'}
tratamento = {'Yes':1, 'No':0}

for i in colunas_dicotomicas:
    df_processado[i] = df_processado[i].map(tratamento)

display(df_processado)
```

3) Transformação das colunas com dados ordinais (hierarquia)

   Colunas: 'work_interfere', 'no_employees', 'benefits', 'care_options', 'wellness_program', 'seek_help', 'anonymity', 'leave', 'mental_health_consequence', 'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview, 'phys_health_interview', 'mental_vs_physical'

```python
escala_work_interfere = {'Never':0, 'Rarely':0.5, 'Sometimes':1.0, 'Often':1.5}
escala_no_employees = {'Micro Business':0, 'Small Business (S1)':0.5, 'Small Business (S2)':1.0, 'Medium-sized Business':1.5, 
                       'Upper-mid Market':2.0, 'Large Enterprise':2.5}
escala_leave = {"Don't know":0, 'Somewhat easy':0.5, 'Very easy':1.0, 'Somewhat difficult':1.5, 'Very difficult':2.0}
escala_geral = {'No':0, "Don't know":0.5, 'Not Sure':0.5, 'Maybe':0.5, 'Some of them':0.5, 'Yes':1.0}


df_processado['work_interfere'].str.strip()
df_processado['work_interfere'] = df_processado['work_interfere'].map(escala_work_interfere)

df_processado['no_employees'].str.strip()
df_processado['no_employees'] = df_processado['no_employees'].map(escala_no_employees)

df_processado['leave'].str.strip()
df_processado['leave'] = df_processado['leave'].map(escala_leave)


colunas_restantes = {'benefits', 'care_options', 'wellness_program', 'seek_help', 'anonymity', 'mental_health_consequence', 
                     'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview', 'mental_vs_physical'}


for i in colunas_restantes:
    df_processado[i] = df_processado[i].map(escala_geral)

display(df_processado)
```

```python
'''
## Percorrimento das colunas para contagem de cada um dos valores e análise
for i in df_processado.columns:
    valores = df_processado[i].value_counts()
    print(valores, "\n")
'''
```

###Criação da matriz de correlação dos atributos

```python
matriz_correlacao = df_processado.corr()

plt.figure(figsize=(24,30))

sns.heatmap(
    matriz_correlacao,
    annot=True,     # Mostra os valores numéricos na célula
    cmap='coolwarm', # Escolha um mapa de cores (ex: 'coolwarm', 'viridis', 'RdBu')
    fmt=".2f",      # Formata os números com 2 casas decimais
    linewidths=.5,  # Linhas entre as células
    cbar=True       # Mostra a barra de cores
)

## Comando de criação do título da figura e comando de exibição na tela ##
plt.title('Mapa de Calor da Matriz de Correlação')
plt.show()
```

###Início do treinamento do modelo

Considerando a matriz gerada, foi realizada a seguinte análise:
1) Atributos com maior correlação com a variável target
2) Atributos com maior quantidade de correlações com outros atributos da matriz
3) Atributos com baixas correlações com todos os demais na matriz (para remoção do treinamento)

Divisão do X e do y

```python
X = df_processado.drop(columns=['Age', 'treatment', 'self_employed', 'remote_work', 'tech_company','wellness_program', 'seek_help', 'anonymity', 'leave',
                       'phys_health_consequence', 'coworkers', 'supervisor', 'mental_health_interview', 'phys_health_interview','mental_vs_physical', 
                       'obs_consequence', 'Gender_FEMALE', 'Gender_MALE','Gender_OTHER', 'Country_Other', 'Country_United_States'])
y = df_processado['treatment']

print(X)
print(y)
```

Divisão entre treino e teste

```python
X_treino, X_teste, y_treino, y_teste = train_test_split(X, y, test_size=0.3, stratify= y, shuffle= True, random_state= None)

print("Formato dos dados de treino", X_treino.shape , " --- " , y_treino.shape)
print("Formato dos dados de teste", X_teste.shape , " --- " , y_teste.shape)
```

Treinamento e teste do modelo

```python
myrf = RandomForestClassifier(n_estimators=100)
myrf.fit(X_treino, y_treino)
score = myrf.score(X_teste, y_teste)

print(score)
```
